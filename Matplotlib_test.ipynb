```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Date = ['Jan.','Feb.','Mar.','Apr.','May','Jun.','Jul.','Aug.','Sep.','Oct.','Nov.','Dec.']
Value = [3.4,3.8,4.0,3.9,3.5,3.6,3.6,3.9,3.8,3.7,3.8,4.0]
plt.plot(Date,Value)
plt.xticks()
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948")
plt.show()
```

```python
from numpy import arange
import matplotlib.pyplot as plt

# Vertical bar plot 
bar_heights = [4.6,5,3.2,4.1,3.6]
num_cols = ['jazeera','google','facebook','apple','sumsung']
# Positions of the left sides of the 5 bars. [0.75, 1.75, 2.75, 3.75, 4.75]
bar_positions = arange(5) + 0.75

tick_positions = range(1,6)

fig, ax = plt.subplots()

ax.bar(bar_positions, bar_heights, 0.5)
ax.set_xticks(tick_positions)
ax.set_xticklabels(num_cols, rotation=90)
ax.set_xlabel('Rating Source')
ax.set_ylabel('Average Rating')
ax.set_title('Average User Rating For top companies')
plt.show()
```

```python
# Horisontal bar plot 
bar_heights = [4.6,5,3.2,4.1,3.6]
num_cols = ['jazeera','google','facebook','apple','sumsung']
# Positions of the left sides of the 5 bars. [0.75, 1.75, 2.75, 3.75, 4.75]
bar_positions = arange(5) + 0.75

tick_positions = range(1,6)

fig, ax = plt.subplots()

ax.barh(bar_positions, bar_heights, 0.5)
ax.set_yticks(tick_positions)
ax.set_yticklabels(num_cols, rotation=90)
ax.set_ylabel('Rating Source')
ax.set_xlabel('Average Rating')
ax.set_title('Average User Rating For top companies')
plt.show()
```

```python
v1 = [1,2,3,4,5]
v2 = [2,3,4,5,6]
fig, ax = plt.subplots()
ax.scatter(v1, v2)
ax.set_xlabel('Variable_1')
ax.set_ylabel('Variable_2')
plt.show()
```

```python
# Scatter plot 
import matplotlib.pyplot as plt

v1 = [1,2,3,4,5]
v2 = [2,3,4,5,6]
v3 = [6,5,4,3,2]
v4 = [2,3,4,5,6]
v5 = [1,2,3,4,5]
v6 = [3,1,2,3,6]
Date = ['Jan.','Feb.','Mar.','Apr.','May','Jun.','Jul.','Aug.','Sep.','Oct.','Nov.','Dec.']
Value = [3.4,3.8,4.0,3.9,3.5,3.6,3.6,3.9,3.8,3.7,3.8,4.0]

fig = plt.figure(figsize=(20,20))
ax1 = fig.add_subplot(6,3,1)
ax2 = fig.add_subplot(6,3,2)
ax3 = fig.add_subplot(6,3,3)
ax4 = fig.add_subplot(6,3,4)
ax5 = fig.add_subplot(6,3,5)
ax6 = fig.add_subplot(6,3,6)


ax1.set_xlim(0, 7)
ax1.set_ylim(0, 7)
ax2.set_xlim(0, 7)
ax2.set_ylim(0, 7)
ax3.set_xlim(0, 7)
ax3.set_ylim(0, 7)


ax1.scatter(v1,v2)
ax1.set_xlabel('Variable_1')
ax1.set_ylabel('Variable_2')

ax2.scatter(v3,v4)
ax2.set_xlabel('Variable_3')
ax2.set_ylabel('Variable_4')

ax3.bar(v5,v6)
ax3.set_xlabel('Variable_5')
ax3.set_ylabel('Variable_6')

# plt.plot(Date,Value)
# plt.xticks(rotation = 90)
# plt.xlabel("Month")
# plt.ylabel("Unemployment Rate")
# plt.title("Monthly Unemployment Trends, 1948")

ax4.plot(Date,Value,linewidth=2, c="r",linestyle='-')
ax4.set_xlabel('Month')
ax4.set_ylabel('Unemployment Rate')

ax5.plot(Date,Value,linewidth=2, c="g",linestyle='--')
ax5.set_xlabel('Month')
ax5.set_ylabel('Unemployment Rate')

ax6.plot(Date,Value,'pr--', linewidth=2, ms=12,mec ='g',mfc = 'g')
ax6.set_xlabel('Month')
ax6.set_ylabel('Unemployment Rate')

plt.show()
```
